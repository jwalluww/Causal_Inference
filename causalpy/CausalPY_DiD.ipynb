# Causalpy QuickStart

```python
import causalpy as cp


# Import and process data
df = (
   cp.load_data("drinking")
   .rename(columns={"agecell": "age"})
   .assign(treated=lambda df_: df_.age > 21)
   )

# Run the analysis
result = cp.pymc_experiments.RegressionDiscontinuity(
   df,
   formula="all ~ 1 + age + treated",
   running_variable_name="age",
   prediction_model=cp.pymc_models.LinearRegression(),
   treatment_threshold=21,
   )

# Visualize outputs
fig, ax = result.plot();

# Get a results summary
result.summary()
```

# Difference in Difference with PYMC models

```python
import causalpy as cp
```

```python
%load_ext autoreload
%autoreload 2
```

```python
df = cp.load_data("did")
df.head()
```

```python
# df['group'].value_counts(dropna=False)
# df['t'].value_counts(dropna=False)
# df['unit'].value_counts(dropna=False)
# df['treated'].value_counts(dropna=False)
df['y'].describe()
```

```python
result = cp.pymc_experiments.DifferenceInDifferences(
    df,
    formula="y~1+group+t+treated:group",
    time_variable_name="t",
    group_variable_name="group",
    treated=1,
    untreated=0,
    prediction_model=cp.pymc_models.LinearRegression(),
)
```

```python
fig,ax = result.plot()
```

```python
result.summary()
```

```python
import arviz as az
ax = az.plot_posterior(result.causal_impact,ref_val=0)
ax.set(title="Posterior estimate of causal impact")
```

# Difference in Differences with scikit learn models

```python
import causalpy as cp
```

```python
data = cp.load_data("did")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
result = cp.skl_experiments.DifferenceInDifferences(
    data,
    formula="y ~ 1 + group + t + treated:group",
    time_variable_name="t",
    prediction_model=LinearRegression(),
)
```

```python
fig, ax = result.plot();
```

# Difference in Differences - Banking Dataset Example

```python
import causalpy as cp
```

```python
%load_ext autoreload
%autoreload 2
```

```python
df = (
    cp.load_data("banks")
    # just keep columns we want
    .filter(items=["bib6", "bib8", "year"])
    # rename to meaningful variables
    .rename(columns={"bib6": "Sixth District", "bib8": "Eighth District"})
    # reduce from daily resolution to examine median banks open by year
    .groupby("year").median()
)
```

```python
ax = df[["Sixth District", "Eighth District"]].plot(style="o-")
ax.set(ylabel="Number of banks in business")
ax.axvline(x=1930.5, c="r", lw=3, label="intervention");
```
